### 3.3 Shift for FFT

```python
w = 0.055 # rect half width (m)
L = 2 # vector side length (m)
M = 200 # No. samples
dx = L/M # sample interval (m)
x = np.linspace(-L/2, L/2-dx, M, endpoint=True) # coordinate vector
f = rect(x / (2*w))
f0 = np.fft.fftshift(f)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 3))
ax = axes.ravel()
ax[0].plot(x, f, '.', markersize=4)
ax[0].plot(x, f, color='red', linewidth=1)
ax[1].plot(x, f0, '.', markersize=4)
ax[1].plot(x, f0, color='red', linewidth=1)
plt.show()
```

### 3.4 Computing the FFT and displaying results

```python
w = 0.055 # rect half width (m)
L = 2 # vector side length (m)
M = 200 # No. samples
dx = L/M # sample interval (m)
x = np.linspace(-L/2, L/2-dx, M, endpoint=True) # coordinate vector
f = np.fft.fftshift(rect(x / (2*w)))
F0 = np.fft.fft(f) * dx
F = np.fft.fftshift(F0)
fx = np.linspace(-1/(2*dx), 1/(2*dx) - 1/L, M, endpoint=True)


fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 6))
ax = axes.ravel()
# ax[0].plot(x, f, '.', markersize=4)
ax[0].plot(abs(F0), color='red', linewidth=1)
# ax[1].plot(x, f0, '.', markersize=4)
ax[1].plot(np.angle(F0), color='blue', linewidth=1)
ax[2].plot(fx, np.abs(F), color='red', linewidth=1)
ax[3].plot(fx, np.abs(F) * np.sign(np.cos(np.angle(np.fft.fftshift(F0)))), color='red', linewidth=1)

plt.show()
```

### 3.5 Comparison with analytic results

```python
w = 0.055 # rect half width (m)
L = 2 # vector side length (m)
M = 200 # No. samples
dx = L/M # sample interval (m)
x = np.linspace(-L/2, L/2-dx, M, endpoint=True) # coordinate vector
f = np.fft.fftshift(rect(x / (2*w)))
F0 = np.fft.fft(f) * dx
F = np.fft.fftshift(F0)
fx = np.linspace(-1/(2*dx), 1/(2*dx) - 1/L, M, endpoint=True)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))
ax = axes.ravel()

# analytic function
F_an = 2*w*np.sinc(2*w*fx)
ax[0].plot(fx, abs(F), 'r', )
ax[0].plot(fx, abs(F_an), 'b.', )
ax[1].plot(fx, np.angle(F), 'r', )
ax[1].plot(fx, np.angle(F_an), 'b.', )
```

### 3.6 Convolution example

```python
## convolution example
wa = 0.3 # Gaussian 1 width [exp(-pi) radius] (m)
wb = 0.2 # Gaussian 2 width [exp(-pi) radius] (m)
L = 2    # side length (m)
M = 200  # No. of samples
dx = L/M # sample interval (m)

x = np.linspace(-L/2, L/2-dx, M, endpoint=True)
fa = np.exp(-np.pi*x**2/wa**2)
fb = np.exp(-np.pi*x**2/wb**2)

Fa = np.fft.fft(np.fft.fftshift(fa))
Fb = np.fft.fft(np.fft.fftshift(fb))
# Fa = np.fft.fft(fa)
# Fb = np.fft.fft(fb)
F0 = Fa*Fb
f0 = np.fft.ifft(F0) * dx
f = np.fft.fftshift(f0)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))
ax = axes.ravel()
# function
ax[0].plot(x, fa, 'r')
ax[0].plot(x, fb, 'b--')

# convolution
ax[1].plot(x, abs(f), 'r')
plt.show()
```

### 3.7 Two dimensions

```python
# 2D example
wx = 0.1   # rect x half-width (m)
wy = 0.05  # rect y half-width (m)
L = 2      # side length x&y (m)
M = 200    # No. sample per side length
dx = L/M   # sample interval (m)
x = np.linspace(-L/2, L/2-dx, M, endpoint=True)
y = x
X, Y = np.meshgrid(x, y)
g = rect(X/(2*wx)) * rect(Y/(2*wy))

g0 = np.fft.fftshift(g)
G0 = np.fft.fft2(g0) * dx**2
G = np.fft.fftshift(G0)
fx = np.linspace(-1/(2*dx), 1/(2*dx)-1/L, M, endpoint=True)
fy = fx


fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 6))
ax = axes.ravel()

ax[0].imshow(g)
ax[0].axis('equal')

ax[1].plot(x, g[int(M/2+1), :])
ax[2].plot(y, g[:, int(M/2+1)])

ax[3].imshow(g0)
ax[4].imshow(np.abs(G))
plt.show()
```

#### 3.2

```python
## convolution example
wa = 0.3 # Gaussian 1 width [exp(-pi) radius] (m)
wb = 0.2 # Gaussian 2 width [exp(-pi) radius] (m)
L = 2    # side length (m)
M = 200  # No. of samples
dx = L/M # sample interval (m)

x = np.linspace(-L/2, L/2-dx, M, endpoint=True)
fa = np.exp(-np.pi*x**2/wa**2)
fb = np.exp(-np.pi*x**2/wb**2)

Fa = np.fft.fft(np.fft.fftshift(fa))
Fb = np.fft.fft(np.fft.fftshift(fb))
F0 = Fa*Fb
f0 = np.fft.ifft(F0) * dx
f = np.fft.fftshift(f0)
analytic_conv = wa*wb/np.sqrt(wa**2 + wb**2) * np.exp(-np.pi*x**2/(wa**2 + wb**2))

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))
ax = axes.ravel()
# function
ax[0].plot(x, fa, 'r')
ax[0].plot(x, fb, 'b--')

# convolution
ax[1].plot(x, abs(f), 'r')
ax[1].plot(x, abs(analytic_conv), 'b--') # analytic convolution

plt.show()
```
