# Module import

```python
from skimage import io
from skimage.transform import rescale, resize, downscale_local_mean
import os, sys
import numpy as np
import matplotlib.pyplot as plt

home_dir = os.path.abspath(os.path.join('..'))
sys.path.append(home_dir)
# if module_path not in sys.path:
# 	sys.path.append(module_path+"\\Common_Module")
from lib.utils import *
from lib.transmittancefunc import *
from lib.propagator import *

%load_ext autoreload
%autoreload 2

def load_usaf(path: str='./USAF-1951.png', size: Tuple[int]=(250, 250)):
    img = io.imread(path)
    img = img[:, :, 3]
    img = resize(img, size, mode='constant', anti_aliasing=False)
    img = np.where(img < 0.99999, 0, 1)
    return img.astype(np.float64)

### define unit
METER = 1
MM = 1e-3 * METER
UM = 1e-6 * METER
NM = 1e-9 * METER
INCH = 25.4 * MM
```

## coh_image coherent imaging example

```python
img = load_usaf()
M, N = img.shape

Ig = img / img.max() # normalize ideal image

ug = np.sqrt(Ig)    # ideal image field
L = 0.3e-3 * METER   # image plane side length
du = L / M           # smaple interval
u = np.linspace(-L/2, L/2 - du, M, endpoint=True)
v = u

wavelength = 0.5e-6 * METER
wxp = 6.25e-3 * METER        # exit pupil radius
zxp = 125e-3 * METER          # exit pupil distance
f0 = wxp / (wavelength * zxp) # cutoff frequency
fu = np.linspace(-1/(2*du), 1/(2*du)-1/L, M, endpoint=True)
fv = fu
Fu, Fv = np.meshgrid(fu, fv)
H = circle_func(Fu, Fv, f0)


params0 = {'set_xlabel' : f'x (um)',
         'set_ylabel' : f'y (um)',
         'set_xticks' : u[::30] / UM,
         'set_yticks' : u[::30] / UM,
         'set_title' : '',
         'extent' : np.array([u[0], u[-1], v[0], v[-1]]) / UM}

params1 = {'set_xlabel' : f'fu, cycles/m',
         'set_ylabel' : f'fv, cycles/m',
         'set_xticks' : fu[::50],
         'set_yticks' : fv[::50],
         'set_title' : '',
         'extent' : np.array([fu[0], fu[-1], fv[0], fv[-1]])}


H = np.fft.fftshift(H)
Gg = np.fft.fft2(np.fft.fftshift(ug))
Gi = H * Gg
ui = np.fft.fftshift(np.fft.ifft2(Gi))
I = np.abs(ui)**2

ug_rough = ug * np.exp(1j * 2 * np.pi * np.random.rand(ug.shape[0], ug.shape[1]))
Gg_rough = np.fft.fft2(np.fft.fftshift(ug_rough))
Gi_rough = H * Gg_rough
ui_rough = np.fft.fftshift(np.fft.ifft2(Gi_rough))
I_rough = np.abs(ui_rough)**2


vvalue = -0.8e-4 * METER
vindex = int(vvalue/du + (M/2+1))

params = [params0, params1, params0, params0, params0]
contents = [img, np.fft.fftshift(H), I**0.5, I, I_rough**0.5]
fig, ax = fig_gen(nrows=2, ncols=3, scale_factor=6)

for idx, (params, content) in enumerate(zip(params, contents)):
    if idx!=3:
        ax[idx] = axis_label(ax[idx], **params)
        ax[idx].imshow(content, extent=params['extent'])
    else:
        ax[idx] = axis_label(ax[idx], **params)
        ax[idx].plot(u / UM, I[vindex, :])
        ax[idx].plot(u / UM, Ig[vindex, :])


plt.show()
plt.show()
```

## incoh_image coherent imaging example

```python
img = load_usaf()
M, N = img.shape

Ig = img / img.max() # normalize ideal image

L = 0.3e-3 * METER   # image plane side length
du = L / M           # smaple interval
u = np.linspace(-L/2, L/2 - du, M, endpoint=True)
v = u

wavelength = 0.5e-6 * METER
wxp = 6.25e-3 * METER        # exit pupil radius
zxp = 125e-3 * METER          # exit pupil distance
f0 = wxp / (wavelength * zxp) # coherent cutoff frequency
fu = np.linspace(-1/(2*du), 1/(2*du)-1/L, M, endpoint=True)
fv = fu
Fu, Fv = np.meshgrid(fu, fv)

H = circle_func(Fu, Fv, f0) # coherent image transfer function

OTF = np.abs(np.fft.fft2(np.fft.fftshift(H)))**2 # optical transfer function
OTF = np.fft.ifft2(OTF)
OTF = np.abs(OTF / OTF[0, 0])

Gg = np.fft.fft2(np.fft.fftshift(Ig))
Gi = Gg * OTF
Ii = np.fft.fftshift(np.fft.ifft2(Gi))

# remove residual image parts, values < 0
Ii = np.where(np.real(Ii) >= 0, np.real(Ii), 0)

params0 = {'set_xlabel' : f'x (um)',
         'set_ylabel' : f'y (um)',
         'set_xticks' : u[::30] / UM,
         'set_yticks' : u[::30] / UM,
         'set_title' : '',
         'extent' : np.array([u[0], u[-1], v[0], v[-1]]) / UM}

params1 = {'set_xlabel' : f'fu, cycles/m',
         'set_ylabel' : f'fv, cycles/m',
         'set_xticks' : fu[::50],
         'set_yticks' : fv[::50],
         'set_title' : '',
         'extent' : np.array([fu[0], fu[-1], fv[0], fv[-1]])}


vvalue = -0.5e-4 * METER
vindex = int(vvalue/du + (M/2+1))

params = [params1, params0, params0]
contents = [np.fft.fftshift(OTF), Ii, Ii[vindex, :]]
fig, ax = fig_gen(nrows=1, ncols=3, scale_factor=6)

for idx, (params, content) in enumerate(zip(params, contents)):
    if idx!=2:
        ax[idx] = axis_label(ax[idx], **params)
        ax[idx].imshow(content, extent=params['extent'])
    else:
        ax[idx] = axis_label(ax[idx], **params)
        ax[idx].plot(u / UM, Ii[vindex, :])
        ax[idx].plot(u / UM, Ig[vindex, :])


plt.show()
plt.show()
```
